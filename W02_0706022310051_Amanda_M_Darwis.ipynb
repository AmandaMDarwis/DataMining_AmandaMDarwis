## Handle Missing Value

```python
minimal_numerical = ['bedrooms', 'bathrooms', 'land_size_m2', 'building_size_m2', 'floors']
for col in minimal_numerical:
    if col in df.columns:
        df[col].fillna(df[col].median(), inplace=True)
# i use median because the missing value of those columns are <1%, and minimalize outlier effects from the data
df.head(20)
```

```python
# Certificate is a caetgorical data, so it can't be easily replaced with median/modus/mean.
df['certificate'].fillna('lainnya', inplace=True)
```

```python
# replacing null with unknown, because the data is categorical, so it can't be just replaced with median or mode
categorical_cols = ['furnishing', 'property_condition', 'building_orientation']

for col in categorical_cols:
    if col in df.columns:
        df[col].fillna('Unknown', inplace=True)
        # print(f"Filled missing values in '{col}' with 'Unknown'")
        # print(f"Value counts for {col}:")
        # print(df[col].value_counts())
        # print("-" * 50)
```

```python
# Fill year_built using median, so it can be used to calculate building_age
df['year_built'].fillna(df['year_built'].median(), inplace=True)

# Fill the missing value in building_age using year_built
# 2022 because from the existing data the current year seems to use 2022 than the current year
df['building_age'] = df.apply(lambda x: 2022 - x['year_built'] if pd.isna(x['building_age']) and not pd.isna(x['year_built']) else x['building_age'], axis=1)

# Now fill the building_age missing value with median (because to minimalize outlier effect)
building_age_median = df['building_age'].median()
df['building_age'].fillna(building_age_median, inplace=True)

# Drop year_built because it's now redundant
df.drop(columns=['year_built'], inplace=True)
```

### Feature Engineering

```python
# Binning price
def bin_price(df):
    # Define bin edges in Rupiah
    bins = [0, 1000000000, 2500000000, 4000000000, 6000000000, float('inf')]

    # Define category labels
    labels = [
        'Under 1B',
        '1B - 2.5B',
        '2.5B - 4B',
        '4B - 6B',
        'Over 6B'
    ]

    # Create binned column
    df['price_category'] = pd.cut(
        df['price_in_rp'],
        bins=bins,
        labels=labels,
        include_lowest=True
    )

    return df

# Apply price binning
df = bin_price(df)

# Show results
print("Price Distribution by Category:")
print(df['price_category'].value_counts().sort_index())

# Visualize
plt.figure(figsize=(10, 6))
df['price_category'].value_counts().sort_index().plot(kind='bar', color='green')
plt.title('Property Price Distribution by Category')
plt.xlabel('Price Category')
plt.ylabel('Number of Properties')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#KEY FINDINGS
# According to the data, the majority of properties are priced under 1 billion.
```

```python
# Binning for land size and building size
def bin_property_sizes(df):

    # Land size binning (in m²)
    land_bins = [0, 80, 120, 180, 250, float('inf')]
    land_labels = [
        'Small (<80m²)',
        'Medium (80-120m²)',
        'Large (120-180m²)',
        'X-Large (180-250m²)',
        'Luxury (>250m²)'
    ]

    # Building size binning (in m²)
    building_bins = [0, 60, 100, 150, 200, float('inf')]
    building_labels = [
        'Compact (<60m²)',
        'Comfortable (60-100m²)',
        'Spacious (100-150m²)',
        'Very Spacious (150-200m²)',
        'Mansion (>200m²)'
    ]

    # Create binned columns
    df['land_size_category'] = pd.cut(
        df['land_size_m2'],
        bins=land_bins,
        labels=land_labels,
        include_lowest=True
    )

    df['building_size_category'] = pd.cut(
        df['building_size_m2'],
        bins=building_bins,
        labels=building_labels,
        include_lowest=True
    )

    return df

# Apply size binning
df = bin_property_sizes(df)

# Show results
print("Land Size Distribution:")
print(df['land_size_category'].value_counts().sort_index())
print("\nBuilding Size Distribution:")
print(df['building_size_category'].value_counts().sort_index())

# Visualize
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Land size
df['land_size_category'].value_counts().sort_index().plot(kind='bar', ax=ax1, color='lightgreen')
ax1.set_title('Land Size Distribution')
ax1.set_xlabel('Land Size Category')
ax1.set_ylabel('Number of Properties')
ax1.tick_params(axis='x', rotation=45)

# Building size
df['building_size_category'].value_counts().sort_index().plot(kind='bar', ax=ax2, color='lightcoral')
ax2.set_title('Building Size Distribution')
ax2.set_xlabel('Building Size Category')
ax2.set_ylabel('Number of Properties')
ax2.tick_params(axis='x', rotation=45)

# plt.tight_layout()
plt.show()


# KEY FINDINGS
# We can see from the land size chart, the majority of properties have land sizes below 80m2
# While the building sizes between 60-100m2 are the most common in this data
```

```python
# Binning building_age
def bin_building_age(df):
    """
    Bin building age into meaningful categories
    """
    # Age binning (in years)
    age_bins = [0, 2, 5, 10, 15, 20, float('inf')]
    age_labels = [
        'Brand New (0-2 years)',
        'Like New (2-5 years)',
        'Recent (5-10 years)',
        'Established (10-15 years)',
        'Mature (15-20 years)',
        'Old (>20 years)'
    ]

    # Create binned column
    df['age_category'] = pd.cut(
        df['building_age'],
        bins=age_bins,
        labels=age_labels,
        include_lowest=True
    )

    return df

# Apply age binning
df = bin_building_age(df)

# Show results
print("Building Age Distribution:")
print(df['age_category'].value_counts().sort_index())

# Visualize
plt.figure(figsize=(10, 6))
df['age_category'].value_counts().sort_index().plot(kind='bar', color='lightblue')
plt.title('Building Age Distribution')
plt.xlabel('Age Category')
plt.ylabel('Number of Properties')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
# Count the facilities of each row
df['facilities_count'] = df['facilities'].apply(lambda x: len(x.split(',')))

# Define the bins and labels based on facilities count
bins = [0, 3, 6, np.inf]
labels = ['1-3 facilities', '4-6 facilities', '>7 facilities']

# Replace the 'facilities' column with the new binned categories
df['facilities'] = pd.cut(df['facilities_count'], bins=bins, labels=labels, right=True)

# Drop the temporary 'facilities_count' column
df = df.drop('facilities_count', axis=1)

df.head()
```
